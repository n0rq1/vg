```python
import pandas as pd
import numpy as np
import networkx as nx
from ts2vg import NaturalVG
import matplotlib.pyplot as plt
```

# Load in dataset 

```python
df = pd.read_csv("./.datasets/BTC-2021min.csv") # load dataset
```

# Create a numpy array of prices

```python
prices = np.array(df["close"]) # create np array of btc prices
```

# Create plot of prices

```python
plt.figure(figsize=(10, 6))
plt.plot(np.arange(len(prices[:5])), prices[:5], color='blue', marker='o', markersize=8, linestyle='-')

plt.xlabel("Index", )
plt.ylabel("Close Price (USD)")
plt.title("BTC Close Prices (First 5)")
plt.xticks(np.arange(len(prices[:5])))
plt.tight_layout()
plt.show()
```

```python
vg = NaturalVG() # initialize VG
vg.build(prices[:5]) # create VG of first 5 prices
G = nx.DiGraph()  # Initialize a directed graph

G.add_nodes_from(vg.as_networkx().nodes())
for i, j in vg.as_networkx().edges():
    if i < j:
        G.add_edge(i, j)  
    else:
        G.add_edge(j, i)
```

# Visualize NVG

```python
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='black', arrows=True, node_size=700, font_size=12)

plt.title("Directed Natural Visibility Graph")
plt.show()
```
